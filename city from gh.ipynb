```python
import requests
import pandas as pd
```

```python
# this forms a dict with level-2 cities as keys and their subordinate units as values 
# this is for matching cases where only lower-level addresses are given
#the data is from https://github.com/adyliu/china_area project (need description on validity)
```

```python
url = "https://raw.githubusercontent.com/adyliu/china_area/master/area_code_2020.json"
df = pd.read_json(url)
```

```python
dct_city={}
```

```python
for row in df.iterrows():
    prov = row[1]['name']
    for dct in row[1]['children']: #only consider city level = 1 and 2       
        if dct['name'] == '市辖区':
            key = row[1]['name']
        elif dct['name'] != '区直辖县级行政区划':
            key = dct['name']
        rel_city = [d['name'] for d in dct['children']]
        rel_city.append(prov)
        dct_city[key] = rel_city
    
```
